```python
#which store has biggest and lowest area
biggest_area = df[(df['Store_Area'] == df['Store_Area'].max())]
lowest_area = df[(df['Store_Area'] == df['Store_Area'].min())]

print (biggest_area)
print (lowest_area)
```

```python

#which store has most and least Daily_Customer_Count
most_customer = df[(df['Daily_Customer_Count'] == df['Daily_Customer_Count'].max())]
least_customer = df[(df['Daily_Customer_Count'] == df['Daily_Customer_Count'].min())]

print (most_customer)
print (least_customer)

```

```python
#which store has most and least Items_Available
most_items = df[(df['Items_Available'] == df['Items_Available'].max())]
least_item = df[(df['Items_Available'] == df['Items_Available'].min())]

print (most_items)
print (least_item)
```

```python
#which store has most and least Store_Sales
most_sales = df[(df['Store_Sales'] == df['Store_Sales'].max())]
least_sales = df[(df['Store_Sales'] == df['Store_Sales'].min())]

print (most_sales)
print (least_sales)
```

```python
# Sort the data by store sales in descending order
sorted_data = df.sort_values('Store_Sales', ascending=False)

# Select the top 10 rows of the sorted data
top_10 = sorted_data.head(10)

# Create a bar chart using pandas
top_10.plot(kind='bar', x='Store_Id', y='Store_Sales', figsize=(10,7))
plt.xlabel('Store ID')
plt.ylabel('Store Sales')
plt.title('Top 10 Stores by Sales')
plt.show()
```

```python
# Sort the data by Daily Customer Count in descending order
sorted_dc = df.sort_values('Daily_Customer_Count', ascending=False)

# Select the top 10 rows of the sorted data
top_10 = sorted_dc.head(10)

# Create a bar chart using pandas
top_10.plot(kind='bar', x='Store_Id', y='Daily_Customer_Count', figsize=(10,7))
plt.xlabel('Store ID')
plt.ylabel('Daily Customer Count')
plt.title('Top 10 Stores by Daily Customer Count')
plt.show()
```
